```python
# Read the data frame
import pandas as pd
messages = pd.read_csv('D:/Projects/NLP/spamhamclassifier.csv', sep=',', names=['label', 'message','host','guest','affrid'], encoding='ISO-8859-1')
```

```python
messages=messages.iloc[:,0:2]
```

```python
# Remove the 0th row
messages = messages.drop(0)

# Reset the index
messages.reset_index(drop=True, inplace=True)
```

```python
messages
```

```python
# Data cleaning and preprocessing 
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
messages.shape
```

```python
y=pd.get_dummies(messages['label'])
y=y.iloc[:,0].values
```

```python
# Train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20,random_state=42)
```

```python
# create the bag of words model and Ngrams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(1,2),binary=True)
```

```python
cv.vocabulary_
```

```python
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
spam_detect_model
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("BOW Model accuracy: ",accuracy_score(y_test,y_pred))
```

```python
print("BOW Model classification_report: ",classification_report(y_test,y_pred))
```

```python
print("BOW confusion matrix: ",confusion_matrix(y_test,y_pred))
```
